#### 2. Handle Missing values

```python
data2 = pd.read_csv('car-sales-missing-data.csv')
```

```python
data2
```

```python
data2.Make = data2.Make.fillna('Tata')
```

```python
data2
```

```python
# using inplace parameter
data2.Doors.fillna(data2.Doors.mean(), inplace=True)
```

```python
data2
```

```python
# using inplace
data2['Price'].fillna('0', inplace=True)
```

```python
data2
```

```python
# drop the missing records
data2.dropna(inplace=True)
```

```python
data2
```
